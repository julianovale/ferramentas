```
import pandas as pd
pd.set_option('display.max_columns',100)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
```

```
url = 'https://raw.githubusercontent.com/julianovale/ferramentas/main/melb_data.csv'
data = pd.read_csv(url)
```

```
cols = ['Price','Landsize','BuildingArea']
for col in cols:
    sns.boxplot(col, data=data)
    plt.show()
```

```
data.describe(include='object')
```

```
for column in data.select_dtypes(include='object'):
    if data[column].nunique() < 10:
        sns.countplot(y=column, data=data)
        plt.show()
```

```
for column in data.select_dtypes(include='object'):
    if data[column].nunique() < 10:
        sns.boxplot(y=column, x='Price', data=data)
        plt.show()
```

```
for column in data.select_dtypes(include='object'):
    if data[column].nunique() < 10:
        display(data.groupby(column).agg(['mean','std']))
```

```
corrs = data.corr()
corrs
```

```
plt.figure(figsize=(10,8))
sns.heatmap(corrs, cmap='RdBu_r', annot=True)
plt.show()
```
